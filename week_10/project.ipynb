```python
import tkinter as tk

class Zenith:
    def __init__(self, name):
        self.name = name

    def unique_services(self):
        pass

    def mutual_services(self):
        return ["Lines of credit", "Investment management and accounts", "Insurance"]

class RetailBanking(Zenith):
    def unique_services(self):
        return ["Retirement and education accounts", "Loans and mortgages", "Checking and saving"]

class GlobalBanking(Zenith):
    def unique_services(self):
        return ["Multi-currency management services and products", "Foreign currency accounts", "Foreign currency credit cards", "Transborder advisory services", "Liquidity management"]

class CommercialBanking(Zenith):
    def unique_services(self):
        return ["Advisory services"]

class GUI:
    def __init__(self, window):
        self.windo = window
        window.title("Zenith Bank Nigeria")

        self.label_name = tk.Label(root, text="Employee Name:")
        self.label_division = tk.Label(root, text="Division:")
        self.entry_name = tk.Entry(root)
        self.entry_division = tk.Entry(root)
        self.btn_submit = tk.Button(root, text="Submit", command=self.display_services)
        self.result_text = tk.StringVar()
        self.result_label = tk.Label(root, textvariable=self.result_text)

        self.label_name.grid(row=0, column=0, sticky="w")
        self.label_division.grid(row=1, column=0, sticky="w")
        self.entry_name.grid(row=0, column=1)
        self.entry_division.grid(row=1, column=1)
        self.btn_submit.grid(row=2, columnspan=2)
        self.result_label.grid(row=3, columnspan=2)

    def display_services(self):
        name = self.entry_name.get()
        division = self.entry_division.get()

        if division.lower() == "retail banking":
            employee = RetailBanking(name)
        elif division.lower() == "global banking":
            employee = GlobalBanking(name)
        elif division.lower() == "commercial banking":
            employee = CommercialBanking(name)
        else:
            self.result_text.set("Invalid Division!")
            return

        unique = employee.unique_services()
        mutual = employee.mutual_services()

        services = "\n".join(unique + mutual)
        self.result_text.set(f"{name} in {division} offers the following services:\n{services}")

def main():
    window = tk.Tk()
    gui = GUI(window)
    window.mainloop()

if __name__ == "__main__":
    main()
```
