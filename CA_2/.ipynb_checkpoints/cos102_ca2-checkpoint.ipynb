```python
Pseudocode for COS102 project:
1. Import the necessary module tkinter as tk

2. Define an empty list to store the orders: order = []

3. Define a function called other_window:
   a. Create a new toplevel window using Toplevel() and configure its properties
   b. Add labels and entries to input name, order, and quantity
   c. Add labels and entries to input whether to make another order
   d. Create "Yes" and "No" buttons for making another order or closing the window
   e. Retrieve the values entered for name, order, and quantity
   f. Based on the entered order, calculate the price and append it to the order list
   g. Run the window

4. Define a function called orders:
   a. Create a new toplevel window with entry widgets for order, quantity, and whether to make another order
   b. Retrieve the values entered for order, quantity, and name
   c. Calculate the price based on the entered order and quantity, and append it to the order list
   d. Run the window

5. Define a function called closed:
   a. Calculate the total price of the orders using the sum of the order list
   b. Determine the discount based on the total price
   c. Display a messagebox with the total price after discount

6. Create the main Tkinter window:
   a. Set the title to "PAU Cafeteria Menu"
   b. Load the food image and create a canvas with scrollbars to display the menu items and prices
   c. Add the various menu items to the canvas to display the menu
   d. Create a button to open the order input window
   e. Set the geometry of the main window to match the size of the food image
   f. Run the window
```
