```python
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk


def submit():
    location = location_entry.get()
    weight = weight_entry.get()
    if location == 1 and weight >= 10:
        messagebox.showinfo(f"Good day, you are expected to pay the sum of N5,000 for your package delivery")
    if location == 1 and weight < 10:
        messagebox.showinfo(f"Good day, you are expected to pay the sum of N3,500 for your package delivery")
    if location == 2 and weight >= 10:
        messagebox.showinfo(f"Good day, you are expected to pay the sum of N10,000 for your package delivery")
    if location == 2 and weight < 10:
        messagebox.showinfo(f"Good day, you are expected to pay the sum of N5,000 for your package delivery")

root = tk. Tk()
root.title("Delivery Request Form")
root.geometry("1080x720")

location_label = tk.Label(root, text="Please input the location you want to deliver the package to: (1 for Ibeju-Lekki and 2 for Epe)")
location_label.pack()
location_entry = tk. Entry(root)
location_entry.pack()

weight_label = tk.Label(root, text="Please input the weight of your package (in kg):")
weight_label.pack()
weight_entry = tk.Entry(root)
weight_entry.pack()

submit_button = tk.Button(root, text="Submit", command=submit)
submit_button.pack()

root.mainloop()
```
